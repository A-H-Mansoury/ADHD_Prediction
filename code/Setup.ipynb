# Data Balance

load data

```python
chunk_size = 1000
chunks = pd.read_csv("/content/drive/MyDrive/DM_project2/full_pearsonr.csv", chunksize=chunk_size)
df_pearsonr = pd.concat(chunks, ignore_index=True)
```

Plot label distributions

```python
df_pearsonr["DX"].value_counts().plot.bar()
```

As we can see our data is highly imbalance in the case that we consider different types of ADHD.

Convert 4 class problem into Binary by aggregating participant with different types of ADHD

```python
df_pearsonr["DX Binary"] = df_pearsonr["DX"].apply(lambda x: 0 if x==0 else 1)
```

Plot distribtion of binary classes

```python
df_pearsonr["DX Binary"].value_counts().plot.pie()
```

As we can see our classes are not that imbalance. Hence, Extra care is not needed in the case of binary classes.

Conclusion, due to the nature of the data. we are unable to augment data to further balance 4 class data. Therefore we will use the binary labels from now on.
